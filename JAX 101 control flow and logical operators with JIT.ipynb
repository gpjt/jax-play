```python
from jax import grad, jit
import jax.numpy as jnp
```

```python
@jit
def f(x):
    for i in range(3):
        x = 2 * x
    return x

print(f(3))
```

```python
@jit
def g(x):
    y = 0.
    for i in range(x.shape[0]):
        y = y + x[i]
    return y

print(g(jnp.array([1., 2., 3.])))
```

```python
@jit
def f(x):
    if x < 3:
        return 3. * x ** 2
    else:
        return -4 * x

try:
    f(2)
except Exception as ex:
    print(ex)
```

```python
@jit
def g(x):
    return (x > 0) and (x < 3)

try:
    g(2)
except Exception as ex:
    print(ex)
```

```python
def python_check_positive_even(x):
    is_even = x % 2 == 0
    return is_even and (x > 0)
```

```python
print(python_check_positive_even(24))
```

```python
@jit
def jax_check_positive_even(x):
    is_even = x % 2 == 0
    return jnp.logical_and(is_even, x > 0)

print(jax_check_positive_even(24))
```

```python
try:
    print(jit(python_check_positive_even)(24))
except Exception as ex:
    print(ex)
```

```python
x = jnp.array([-1, 2, 5])
print(jax_check_positive_even(x))
```

```python
try:
    print(python_check_positive_even(x))
except Exception as ex:
    print(ex)
```
